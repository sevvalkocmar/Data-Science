```python
import re
for index, row in final.iterrows():
    reString = re.findall(r'\d+x\d+',row.ScreenResolution)
    splited = reString[0].split('x')
    final.at[index, 'Width'] = (int)(splited[0])
    final.at[index, 'Height'] = (int)(splited[1])
final
```

# Veri Düzenleme

```python
final['Ram'].replace(r'GB','', regex=True, inplace=True)
final['Weight'].replace(r'kg','', regex=True, inplace=True)

final["Ram"] = pd.to_numeric(final["Ram"])
final["Weight"] = pd.to_numeric(final["Weight"])
final.dtypes
```

# MEMORY sütun düzenlemesi

```python
final1=final.copy()
final1["SSD"]='0'
final1["HDD"]='0'
final1["FLASH"]='0'
final1["HYBRID"]='0'
final1
```

```python
for index, row in final1.iterrows():
    splitted = row.Memory.split('+')
    for value in splitted:
        if 'SSD' in value:
            final1.at[index,'SSD'] = value
        elif 'HDD' in value:
            final1.at[index,'HDD'] = value
        elif 'Flash Storage' in value:
            final1.at[index,'FLASH']= value
        else:
            final1.at[index,'HYBRID'] = value
```

```python
final1
```

```python
final1.SSD.replace(r'SSD','', regex=True, inplace=True)
final1.HDD.replace(r'HDD','', regex=True, inplace=True)
final1.FLASH.replace(r'Flash Storage','', regex=True, inplace=True)
final1.HYBRID.replace(r'Hybrid','', regex=True, inplace=True)

final1
```

```python
final1.SSD.replace(r'GB','', regex=True, inplace=True)
final1.HDD.replace(r'GB','', regex=True, inplace=True)
final1.FLASH.replace(r'GB','', regex=True, inplace=True)
final1.HYBRID.replace(r'GB','', regex=True, inplace=True)

```

```python

def clearTB(x):
    if "TB" in x:
        x = x.replace("TB", "")
        #.replace(r'TB','', regex=True, inplace=True)
        x = ((float)(x)) * 1024
        x = ((str)(x))
        return x
    else:
        return x

final1['HDD'] = pd.to_numeric(final1['HDD'].apply(clearTB))
final1['SSD'] = pd.to_numeric(final1['SSD'].apply(clearTB))
final1['FLASH'] = pd.to_numeric(final1['FLASH'].apply(clearTB))
final1['HYBRID'] = pd.to_numeric(final1['HYBRID'].apply(clearTB))
```

```python
final2 = final1.copy();
final1
```
